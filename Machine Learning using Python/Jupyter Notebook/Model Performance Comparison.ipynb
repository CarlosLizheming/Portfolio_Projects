```python
import matplotlib.pyplot as plt
```

### Input the best performance for each fine-tuned model

```python
model_name=['Logistic Regression','SVM','Random Forest','Naive Bayes']
f1_macro_avg=[0.75,0.84,0.89,0.60]
f1_macro_weg=[0.87,0.91,0.94,0.59]
f1_class_3  =[0.52,0.56,0.67,0.33]
f1_class_5  =[0.45,0.83,0.90,0.28]
f1_class_6  =[0.00,0.46,0.53,0.36]
recall_class_6 =[0.00,0.46,0.42,0.92]
precision_class_6 =[0.00,0.46,0.71,0.22]
```

### Class 6 Performance Comparison 

```python
plt.figure(figsize=(8,6))
plt.plot(model_name, f1_class_6,
         label='Class 6 F1 Score',
         color='cornflowerblue', linewidth=2)

plt.plot(model_name, precision_class_6,
         label='Class 6 Precision',
         color='darkorange', linewidth=2)

plt.plot(model_name, recall_class_6,
         label='Class 6 Recall',
         color='deeppink', linewidth=2)


plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Models',fontsize=15)
plt.ylabel('Classification Score',fontsize=15)
plt.title('Classification Score Comparison among Models',fontsize=15)
plt.legend(loc="lower left")
plt.show()
```

### F1 Score Comparison

```python
# Plot all ROC curves
plt.figure(figsize=(8,6))
plt.plot(model_name, f1_macro_avg,
         label='macro-average F1 Score',
         color='darkorange', linewidth=2)
for a,b in zip(model_name, f1_macro_avg): 
    plt.text(a, b, str(b), va='top')
    
plt.plot(model_name, f1_macro_weg,
         label='weighted-average F1 Score',
         color='navy', linewidth=2)
for a,b in zip(model_name, f1_macro_weg): 
    plt.text(a, b, str(b), va='bottom')
    
plt.plot(model_name, f1_class_3,
         label='Class 3 F1 Score',
         color='deeppink', linewidth=2, linestyle=':')

    
plt.plot(model_name, f1_class_5,
         label='Class 5 F1 Score',
         color='aqua', linewidth=2, linestyle=':')

    
plt.plot(model_name, f1_class_6,
         label='Class 6 F1 Score',
         color='cornflowerblue', linewidth=2, linestyle=':')

# colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])


# plt.xlim([0.0, 1.0])
# plt.ylim([0.0, 1.05])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Models',fontsize=15)
plt.ylabel('F1-Score',fontsize=15)
plt.title('F1-Score Comparison among Models',fontsize=15)
plt.legend(loc="lower left")
plt.show()
```
